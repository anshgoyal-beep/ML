```python
import math
import random
N = 10000
d = 10
ads_selected =[]
number_of_times1 = [0]*d
number_of_times0 = [0]*d
total_reward = 0
for n in range(0,10000):
    ad = 0
    max_random = 0
    for i in range(0,10):
        random_beta = random.betavariate(number_of_times1[i]+1,number_of_times0[i]+1)
        if random_beta > max_random:
            max_random = random_beta
            ad = i
    ads_selected.append(ad)
    reward  = df.values[n,ad]
    if reward == 1:
        number_of_times1[ad] = number_of_times1[ad]+ 1
    else:
        number_of_times0[ad] = number_of_times0[ad] + 1
    total_reward = total_reward + reward
        
```
